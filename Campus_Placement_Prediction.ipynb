**Data Preprocessing**

```python
data.head()
```

```python
# Lets drop unwanted columns
data= data.drop(['sl_no','salary'],axis=1)
```

```python
data.head()
```

**Encoding The Categorical Columns**

```python
data['ssc_b'].unique()  # Checking the unique columns
```

```python
data['ssc_b'] = data['ssc_b'].map({'Central':1,'Others':0})
```

```python
data.head()
```

```python
data['hsc_b'].unique()  # Checking the unique values
```

```python
data['hsc_b'] = data['hsc_b'].map({'Central':1,'Others':0})
```

```python
data.head()
```

```python
data['hsc_s'].unique()
```

```python
data['hsc_s']  = data['hsc_s'].map({'Commerce':1,'Science':2, 'Arts':0})
```

```python
data.head()
```

```python
data['degree_t'].unique()
```

```python
data['degree_t'] = data['degree_t'].map({'Sci&Tech':2, 'Comm&Mgmt':1, 'Others':0})
```

```python
data.head()
```

```python
data['specialisation'].unique()
```

```python
data['specialisation'] = data['specialisation'].map({'Mkt&HR':1, 'Mkt&Fin':0})
```

```python
data.head()
```

```python
data['workex'].unique()
```

```python
data['workex'] = data['workex'].map({'No':0, 'Yes':1})
```

```python
data.head()
```

```python
data['status'].unique()
```

```python
data['status'] = data['status'].map({'Placed':1, 'Not Placed':0})
```

```python
data.head()
```

**Store Feature Matrix in X and Target in Vector Y**

```python
X = data.drop('status',axis=1)
```

```python
y = data.status
```

```python
X
```

```python
y
```

**Splitting the Data into Training Set and Testing Set**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

**Import Models**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

**Model Training**

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
gb = GradientBoostingClassifier()
gb.fit(X_train,y_train)
```

**Prediction on Test Data**

```python
from pandas.core.arrays.sparse import dtype
y_pred1 = lr.predict(X_test)
y_pred2 = knn.predict(X_test)
y_pred3 = dt.predict(X_test)
y_pred4 = rf.predict(X_test)
y_pred5 = gb.predict(X_test)
```

**Evaluating the Algorithims**

```python
from sklearn.metrics import accuracy_score
```

```python
acc1 = accuracy_score(y_test,y_pred1)
acc2 = accuracy_score(y_test,y_pred2)
acc3 = accuracy_score(y_test,y_pred3)
acc4 = accuracy_score(y_test,y_pred4)
acc5 = accuracy_score(y_test,y_pred5)
```

```python
print(acc1,acc2,acc3,acc4,acc5)
```

```python
df1 = pd.DataFrame({'Models':['LR','KNN','DT','RF','GB'],
                    'ACC':[acc1,acc2,acc3,acc4,acc5]})
```

```python
df1
```

**From the above dataframe as we can check that LR Logistic Regression is the best model for this data .Hence we will take LR as our best model.**
